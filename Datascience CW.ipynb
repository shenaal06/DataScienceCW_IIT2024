```python
import pandas as pd
```

```python
#Loading the dataset
df_1 = pd.read_excel("Global Superstore Lite (1).xlsx")
```

```python
#Explore the structure of the dataset
#View the first few rows
print(df_1.head())
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

# Select relevant columns for Market Basket Analysis
df_subset = df_1[['Order ID', 'Sub-Category','Product Name']]

# Convert 'Order ID' and 'Product Name' columns to string type
df_subset['Order ID'] = df_subset['Order ID'].astype(str)
df_subset['Sub-Category'] = df_subset['Sub-Category'].astype(str)

# Convert the transaction data into a one-hot encoded DataFrame
one_hot_encoded = pd.get_dummies(df_subset['Sub-Category']).groupby(df_subset['Order ID']).max()

# Perform Market Basket Analysis using Apriori algorithm
frequent_itemsets = apriori(one_hot_encoded, min_support=0.001, use_colnames=True)

# Generate association rules with adjusted parameters
rules_2 = association_rules(frequent_itemsets, metric="lift", min_threshold=0)

# Display the association rules
print("Association Rules:\n", rules_2)
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

# Select relevant columns for Market Basket Analysis
df_subset = df_1[['Order ID', 'Sub-Category','Product Name']]

# Convert 'Order ID' and 'Product Name' columns to string type
df_subset['Order ID'] = df_subset['Order ID'].astype(str)
df_subset['Sub-Category'] = df_subset['Sub-Category'].astype(str)

# Convert the transaction data into a one-hot encoded DataFrame
one_hot_encoded = pd.get_dummies(df_subset['Sub-Category']).groupby(df_subset['Order ID']).max()

# Perform Market Basket Analysis using Apriori algorithm
frequent_itemsets = apriori(one_hot_encoded, min_support=0.001, use_colnames=True)

# Generate association rules with adjusted parameters
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Display the association rules
print("Association Rules:\n", rules)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a pivot table of association rules
pivot_table = rules.pivot(index='antecedents', columns='consequents', values='lift')

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_table, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Association Rules Heatmap')
plt.xlabel('Consequents')
plt.ylabel('Antecedents')
plt.show()
```
